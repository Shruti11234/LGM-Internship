```python
#importing libraries
import cv2
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Desktop\WhatsApp Image 2022-03-07 at 10.53.46.jpeg")
```

```python
cv2.imshow("Desktop\WhatsApp Image 2022-03-07 at 10.53.46.jpeg", image)
```

```python
%matplotlib inline
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("New pic", gray_image)
```

```python
plt.imshow(gray_image)
```

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
plt.imshow(inverted_image)
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
plt.imshow(pencil_sketch)
```

```python
def dodgeV2(X,Y):
    return cv2.divide(X,255-Y,scale=256)
```

```python
pencil_sktech=dodgeV2(gray_image, blurred)
plt.imshow(pencil_sktech)
```
